```python
import requests
from bs4 import BeautifulSoup
import datefinder

def ft_news(ticker):
    ft = []
    response = requests.get("http://chstocksearch.herokuapp.com/api/"+str(ticker))
    print(str(response.json()[0]['company']))
    url = "https://www.ft.com/search?q=" +str(response.json()[0]['company']) + "&page=2"
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"html.parser")
    links = soup.find_all("a")
    for link in links:
        print(link)
        if str(link['href'])[0:31] == 'http://www.bloomberg.com/news/a':
            print(link['href'])
ft_news('GOOGL')
    
def get_benzinga_news(ticker):
    benzinga=[]
    url = "http://www.benzinga.com/stock-articles/" + str(ticker) + "/analyst-ratings?page=1"
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    links = soup.find_all("a")
    for link in links:
        try:
            if (str(link['href'][0:34])) == '/analyst-ratings/analyst-color/16/':
                article = {}
                newurl = "http://www.benzinga.com" + str(link['href'])
                req = requests.get(newurl)
                newsoup = BeautifulSoup(req.text, "html.parser")
                newtext = newsoup.find_all("div",class_='article-content-body')
                for text in newtext:
                    if len(text.text) > 0:
                        article['text'] = text.text
                        article['url'] = newurl
                        newlink = newsoup.find("span",class_='date')
                        matches = datefinder.find_dates(str(newlink))
                        for match in matches:
                            article['date'] = str(match)[0:11]
                        benzinga.append(article)
        except KeyError:
            print("err")
    return(benzinga)
        
def get_fidelity_news(ticker):
    fidelity = []
    url = "https://eresearch.fidelity.com/eresearch/evaluate/news/basicNews.jhtml?symbols=" + str(ticker) + "&anchor=1_201609160852MRKTWTCHNEWS_SVC000179"
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    links = soup.find_all("a")
    for link in links:
        try:
            if str(link['href'])[8:18] == 'ch/evaluat':
                article = {}
                req = requests.get("https://eresearch.fidelity.com" + str(link['href']))
                article['url'] = "https://eresearch.fidelity.com" + str(link['href'])
                newsoup = BeautifulSoup(req.text, "html.parser")
                newtext = newsoup.find_all("div",id='articleContainer')
                for text in newtext:
                    article['text'] = text.text
                newlink = newsoup.find("div",class_="byline")
                matches = datefinder.find_dates(str(newlink))
                for match in matches:
                    article['date'] = str(match)[0:11]
                fidelity.append(article)
        except KeyError:
            print("link")
    return(fidelity)
        
```
